```
import pandas as pd
import matplotlib.pyplot as plt

df_ds = pd.read_csv('https://cocl.us/datascience_survey_data')
df_ds.rename(columns={'Unnamed: 0':'Field'}, inplace=True)
df_ds.set_index('Field', inplace=True)
df_ds.sort_values('Very interested', ascending = False, inplace=True)

# Formatting to a percentage will be done in the plotting
df_dsp = df_ds / 2233

#PLOTTING
ax1 = df_dsp.plot(kind = 'bar',
                figsize = (20,7),
                width = 0.8,
                color = ('#5cb85c', '#5bc0de', '#d9534f'),
                fontsize = 14)
ax1.set_title("Percentage of Respondents' Interest in Data Science", 
             fontsize = 16)
ax1.legend(fontsize = 14)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)

# Adding the percentage values    
for p in ax1.patches:
    ax1.annotate("{:.2%}".format(p.get_height()),
                xy=(p.get_x()+0.02, p.get_height()+0.01))
```
